```python
# now we look adr or prices on basis of cancelled data and not cancelled data.
cancelled_data = Hotel_df[Hotel_df['is_canceled'] == 1]
cancelled_adr = cancelled_data.groupby('reversation_status_date')[['adr']].mean()
cancelled_adr.reset_index(inplace = True)
cancelled_adr.sort_values('reversation_status_date',inplace = True)

not_cancelled_data = Hotel_df[Hotel_df['is_canceled'] == 0]
not_cancelled_adr = not_cancelled_data.groupby('reversation_status_date')[['adr']].mean()
not_cancelled_adr.reset_index(inplace = True)
not_cancelled_adr.sort_values('reversation_status_date',inplace = True)

plt.figure(figsize = (15,7))
plt.title('Average Daily Rate',size =30)
plt.plot(not_cancelled_adr['reversation_status_date'],not_cancelled_adr['adr'],color = 'w', label = 'not cancelled')
plt.plot(cancelled_adr['reversation_status_date'],cancelled_adr['adr'],color = 'r',label = 'cancelled')
plt.legend(fontsize=20)
```

### we can see its little messy data and inconsistency in data because of our data set but there is consistency between 2016 to 2017 we see that cancellation rate is high because of prices and some spikes we assume it is in weekends and holidays.

```python
cancelled_adr = cancelled_adr[(cancelled_adr['reversation_status_date']>'2016') & (cancelled_adr['reversation_status_date']<'2017-09')]
not_cancelled_adr = not_cancelled_adr[(not_cancelled_adr['reversation_status_date']>'2016') & (not_cancelled_adr['reversation_status_date']<'2017-09')]

plt.figure(figsize=(15,7))
plt.title('Average Daily Rate' ,fontsize= 30)
plt.plot(not_cancelled_adr['reversation_status_date'],not_cancelled_adr['adr'],color = 'w',label = 'not cancelled')
plt.plot(cancelled_adr['reversation_status_date'],cancelled_adr['adr'],color = 'r',label = 'cancelled')
plt.xlabel('Reservation Status Date',size =20)
plt.ylabel('ADR Of Hotel',size=20)
plt.legend(fontsize=20)
```

```python
# now lets see how much customer paying for one night 
resort_hotel = Hotel_df.loc[(Hotel_df['hotel'] == 'Resort Hotel') & (Hotel_df['is_canceled'] == 0)]
city_hotel = Hotel_df.loc[(Hotel_df['hotel'] == 'City Hotel') & (Hotel_df['is_canceled'] == 0)]
```

```python
# Counting adults and children as paying customers only, not babies.
resort_hotel['adr_pp'] = resort_hotel['adr'] / (resort_hotel['adults'] + resort_hotel['children'])
city_hotel['adr_pp'] = city_hotel['adr'] / (city_hotel['adults'] + city_hotel['children'])

resort_hotel_filtered = resort_hotel.replace([np.inf, -np.inf], 0)
city_hotel_filtered = city_hotel.replace([np.inf, -np.inf], 0)
```

```python
#printing out the prices
print("""From all non-canceled bookings, across all room types and meals, the average prices are:
Resort hotel: {:.2f} $ per night and person.
City hotel: {:.2f} $ per night and person."""
      .format(resort_hotel_filtered['adr_pp'].mean(), city_hotel_filtered['adr_pp'].mean()))
```

```python
# normalize price per night (adr):

Hotel_df['adr_pp'] = Hotel_df['adr'] / (Hotel_df['adults'] + Hotel_df['children'])
Hotel_df_guests = Hotel_df.loc[Hotel_df['is_canceled'] == 0]
Hotel_df_guests
```

```python
# take out room prices according to room type

room_prices = Hotel_df_guests[['hotel', 'reserved_room_type', 'adr_pp']].sort_values('reserved_room_type')
room_prices
```

```python
# boxplot:
plt.figure(figsize=(16, 10))
c = 'Greys'
sns.boxplot(x='reserved_room_type',y='adr_pp',hue='hotel',data = room_prices,hue_order=['City Hotel', 'Resort Hotel'],palette = c)
plt.title('Price of room types per person and night', fontsize=30)
plt.xlabel('Room type', fontsize=20)
plt.ylabel('Price [Dollar]', fontsize=20)
plt.legend(loc='upper right',fontsize=17)
plt.ylim(0, 160)
plt.show()
```
